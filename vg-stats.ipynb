# Video Game Sales 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

## Open And Print The .CSV File

```python
vgsale = pd.read_csv('Game/vgsales.csv')
vgsale
```

## Which company is the most common video game publisher?

```python
most_pub = vgsale['Publisher'].mode()
most_pub[0]
```

## Whatâ€™s the most common platform?

```python
most_com_plat = vgsale['Platform'].mode()
most_com_plat[0]
```

## What about the most common genre?

```python
most_com_gen = vgsale['Genre'].mode()
most_com_gen[0]
```

## What are the top 20 highest grossing games?

```python
top_20 = vgsale[['Name','Global_Sales','NA_Sales']].sort_values(by='Global_Sales',ascending=False).head(20)
top_20
```

## For North American video game sales, whatâ€™s the median?
* Provide a secondary output showing ten games surrounding the median sales output.
   * assume that games with same median value are sorted in descending order.

```python
na_sales = vgsale['NA_Sales'].median()
na_sales
# x= vgsale[vgsale['Name']=='Spider-Man: Edge of Time']
# x
```

```python
median = vgsale[['NA_Sales']].median()[0]
result = vgsale[vgsale['NA_Sales']==median]
leng = len(result)
result = result[(leng//2)-5:(leng//2)+5]
result
```

## For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python
top_selling_game = top_20['NA_Sales'][0]
mean = vgsale['NA_Sales'].mean()
standard_deviation = vgsale['NA_Sales'].std()
result = (top_selling_game - mean) / standard_deviation
mean
top_selling_game
result
```

## The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?

```python
nintendo_average_sales = vgsale[vgsale['Platform']=='Wii']['Global_Sales'].mean()
other_platforms =  vgsale[vgsale['Platform'] != 'Wii']['Global_Sales'].mean()
nintendo_average_sales , other_platforms
```

## Sort The Platforms From Highest Average Sales.

```python
highest_av_sl = vgsale.groupby('Platform').mean().sort_values('Global_Sales',ascending=False)['Global_Sales']
highest_av_sl
```

## Which game has the highest sales in EU?

```python
highest_EU_sl = vgsale.sort_values('EU_Sales',ascending=False).head(1)
highest_EU_sl
```

## Which game has the highest sales in JP?

```python
highest_JP_sl = vgsale.sort_values('JP_Sales',ascending=False).head(1)
highest_JP_sl
```
